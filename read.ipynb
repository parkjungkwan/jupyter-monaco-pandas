import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
 pd.DataFrame([[score() for i in range(1,5) ] for i in range(1,11)], 
              index=[id() for i in range(1, 11)], 
              columns=['국어','영어','수학','사회'])
```

```python
df = pd.DataFrame([[score() for i in range(1,5) ] for i in range(1,11)], 
                  index=[id() for i in range(1, 11)], 
                  columns=['국어','영어','수학','사회'])
```
